```python
describe_df = pd.DataFrame()
for i in df['country'].unique():
    for j in df[df['country'] == i]['product'].unique():
        test = df[(df['country'] == i) & (df['product'] == j)].describe()
        test.insert(0, "label", str(i) + '_' + str(j))
        describe_df = describe_df.append(test)
```

```python
df_chile = df[df['country'] == 'Chile']
df_mexico = df[df['country'] == 'Mexico']
df_colombia = df[df['country'] == 'Colombia']
```

```python
df_chile_final = df_chile.query('product == ["Inno Gialuron", "Varikosette", "Motion Free", "Rechiol", "Duston Gel", "Yaconia"]')
df_mexico_final = df_mexico.query('product == ["Profiber Hermuno", "Biadit", "Leveren", "Rechiol", "Caro Crystall"]')
df_colombia_final = df_colombia.query('product == ["Motion Free", "Rechiol"]')
```

```python
df_mexico_final
```

# Chile

```python
for i in df_chile_final['product'].unique():
    corr_df = df_chile_final[(df_chile_final['product'] == i)]
    corr_df['norm_marketplaces_lead_qty'] = (corr_df['marketplaces_lead_qty']-corr_df['marketplaces_lead_qty'].min())/(corr_df['marketplaces_lead_qty'].max()-corr_df['marketplaces_lead_qty'].min())
    corr_df['norm_lead_qty'] = (corr_df['lead_qty']-corr_df['lead_qty'].min())/(corr_df['lead_qty'].max()-corr_df['lead_qty'].min())
    df2 = corr_df
    df2 = df2.set_index('date')
    df2 = df2.sort_values(by='date')
    
    plt.figure(figsize=(10, 7))

    plt.plot(df2.index, df2['norm_marketplaces_lead_qty'].rolling(7).mean())
    plt.plot(df2.index, df2['norm_lead_qty'].rolling(7).mean())


    plt.title('Динамика лидов во времени, Chile, ' + i)
    plt.xlabel('Месяц')
    plt.ylabel('Нормированное кол-во лидов')
    plt.legend(['marketplaces_lead_qty', 'lead_qty'])

    plt.xticks(df2.month.unique(), rotation=40)

    plt.grid()
    plt.show()
    
    plt.savefig('Chile_'+i+'.pdf')
```
